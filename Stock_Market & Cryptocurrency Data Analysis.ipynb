```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
assets= ["AAPL","MSFT","TSLA","MMM"]
```

```python
df= yf.download(assets, start= "2020-01-01")
```

```python
df
```

```python
prices= df["Close"]
```

```python
prices
```

```python
prices.plot()
```

```python
returns=prices.pct_change()
```

```python
returns
```

```python
cum_ret= (returns+1).cumprod()-1
```

```python
cum_ret
```

```python
cum_ret.plot(figsize=(10,8))
```

```python
returns.std()
```

```python
returns["TSLA"].hist(bins=30)
```

```python
fig,ax= plt.subplots(2,2, figsize=(10,10))

counter= 0
for i in range(2):
    for j in range(2):
        ax[i,j].plot(returns[returns.columns[counter]], color="blue")
        ax[i,j].set_title(returns.columns[counter])
        counter+=1
```

```python
fig,ax= plt.subplots(2,2, figsize=(10,10))

counter= 0
for i in range(2):
    for j in range(2):
        ax[i,j].hist(returns[returns.columns[counter]], bins=50, color= "red")
        ax[i,j].set_title(returns.columns[counter])
        counter+=1
```
